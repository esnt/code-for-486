This function isn't super flexible, but hopefully it will be useful to help you understand how the hyperparameters work. 



You can change the following arguments:

* Data generation parameters:
    * `dataset`:  Can be "blobs", "moons", or "circles".  These come from the scikit-learn toy datasets.
    * `noise`: (default is 0.2).  This controls how much noise is in the data.  Keep this value between 0.1 and 0.4. 
    * `n_samples`: (default is 1000). Sample size.
    * `random_state`: (default is 307).

* Model hyperparameters:
    * `kernel`: (default is 'rbf').  Common kernels are "linear", "poly", "rbf". See [sklearn.svm.SVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html) for list of possible kernels and [kernel functions documentation](https://scikit-learn.org/stable/modules/svm.html#svm-kernels) for definition of kernels.  (*At this time, the "poly" kernel doesn't seem to be plotting correctly.*)
    * `C`: (default is 1.0). Regularization parameter.  The strength of regularization is inversely proportional to C. 
    * `gamma`: (default is "scale"). Kernel coefficient.  See [sklearn.svm.SVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html). *The default of `gamma="scale"` means that 1 / (n_features * X.var()) is used as the value for gamma*.
    * `degree`: (default is 3). Degree of polynomial kernel function.  Ignored by all kernels except "poly".
